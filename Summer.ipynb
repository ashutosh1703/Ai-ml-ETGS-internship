# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd 
df = pd.read_csv("summer.csv")
len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
gold= df[df['Medal']=='Gold']
data = []
for sport in gold['Sport'].unique():
    data.append([sport, len(gold[gold['Sport']==sport])])
data = pd.DataFrame(data, columns=['Sport', 'Gold'])
data.sort_values(by='Gold', ascending=False).head().plot(x = 'Sport', y = 'Gold', kind = 'bar', figsize = (5,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data=[]
for Medal in df['Sport'].unique():
    data.append([Medal, len(df[df['Sport'] == Medal])])
data = pd.DataFrame(data, columns = ['Sport','Medals'])
data.sort_values(by = 'Medals', ascending = False).head().plot(x = 'Sport', y = 'Medals', kind = 'bar', figsize = (5,5))
```

### 4. Which player has won most number of medals? (Top 5)

```python
data=[]
for Medal in df['Athlete'].unique():
    data.append([Medal, len(df[df['Athlete'] == Medal])])
data = pd.DataFrame(data, columns = ['Athlete','Medals'])
data.sort_values(by = 'Medals', ascending = False).head().plot(x = 'Athlete', y = 'Medals', kind = 'bar', figsize = (5,5))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
gold= df[df['Medal']=='Gold']
data = []
for Athlete in gold['Athlete'].unique():
    data.append([Athlete, len(gold[gold['Athlete']==Athlete])])
data = pd.DataFrame(data, columns=['Athlete', 'Gold'])
data.sort_values(by='Gold', ascending=False).head().plot(x = 'Athlete', y = 'Gold', kind = 'bar', figsize = (5,5))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
gold=df[df['Medal']=="Gold"]
gold[gold['Country']=="IND"]['Year'].min()
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
df.groupby("Event").count()['Athlete'].sort_values(ascending=False).head(5).plot.bar(figsize=(15, 5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data=[]
gold=df[df['Medal']=="Gold"]
goldw = gold[gold['Gender']=="Women"]
for sport in goldw['Sport'].unique():
    data.append([sport, len(goldw[goldw['Sport']==sport])])
data = pd.DataFrame(data, columns=['Sport', 'GoldWomen'])
data.sort_values(by='GoldWomen', ascending=False).head().plot(x = 'Sport', y = 'GoldWomen', kind = 'bar', figsize = (10,5))
```
